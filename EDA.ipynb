```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
```

# Feature

```python
data = np.load('data/test/A-1.npy')
df = pd.DataFrame(data)
len(df)
```

```python
fig,axs = plt.subplots(nrows=25, ncols=1, figsize=(10, 30))
axs_flat = axs.flatten()
for i in range(25):
    axs_flat[i].plot(df.iloc[:,i])
    axs_flat[i].set_title(f"Feature {i}")
plt.tight_layout() 
```

# Label

```python
label_df = pd.read_csv('data/labeled_anomalies.csv')
label_df['anomaly_sequences'] = label_df['anomaly_sequences'].apply(lambda x: np.array(eval(x)))
anomaly_index = label_df[label_df['chan_id']=='A-1']
anomaly_index
```

```python
fig,axs = plt.subplots(nrows=25, ncols=1, figsize=(10, 25))
axs_flat = axs.flatten()
anomaly_range = anomaly_index.anomaly_sequences.values[0][0]
for i in range(25):
    axs_flat[i].plot(df.iloc[:,i])
    axs_flat[i].plot(df.iloc[anomaly_range[0]:anomaly_range[1]+1,i],c='tomato')
plt.tight_layout() 
```
